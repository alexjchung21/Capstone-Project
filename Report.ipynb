## Data

To pursue this research project, we will be using the Foursquare API in order to collect data regarding the availability of Asian food places and restaurants in a few prominent locations. We will be taking data regarding the 5 major locations Boston, New York, Chicago, San Francisco, and Dallas. There is a specific API call to only gather the Asian restaurants in a particular area, and we will make 50 calls for each of the 5 locations, making 250 calls in total. Note that we are limited by the number of calls we can make using the Foursquare API. Collecting this data, we will determine the relative density of Asian restaurants in these locations and calculate the average distance to a central point to find how close everything is clustered together for accessibility. 

## Methodology

From the above analysis, we can gather a rudimentary view of the relative density and availability of Asian dining options in the 5 major cities Boston, New York, Chicago, San Francisco, and Dallas. It seems that Boston and New York have a relatively higher density, whereas Dallas' Asian options are a little more spread out. However, we have not yet garnered a numerical indicator for the density - we will do so by fixating on a particular major landmark in the midst of the area, and calculate the average distance from that point to the available Asian cuisines. To do so, we will find the latitude and longitude coordinates of the major landmark in each major city, and then do a simple distance formula combined with the .mean invocation to calculate the mean distance. Doing this will tell us how close local options are, and available they are. Furthermore, we will employ a clustering Kmeans algorithm on each city. By grouping the available locations into different clusters, we can recommend a particular subset of restaurants to tourists who are currently in a certain location and looking for a place to eat.

## Discussion

While our rudimentary calculation was fairly useful in determining the relative densities of Asian food options in the five major cities, there are of courses improvements that can be made. For example, this kind of calculation is rather sensitive to the presence of outliers. For example, if there happen to be a few points which are far away from the others, this drastically 
increases the average/mean distance. Thus, it might benefit us to take away these few points, because chances are that the tourist isn't lacking choice to choose from; eliminating one or two outliers doesn't mean he has nowhere to eat as there are still around 48 points in each dataset. Based on our results, we can make the recommendation that Boston, MA and New York, NY and San Francisco, CA are among the best places to go for Asian dining while being able to visit major landmarks and famous places in the city.
